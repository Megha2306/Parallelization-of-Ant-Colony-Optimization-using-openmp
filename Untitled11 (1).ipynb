```python
import matplotlib.pyplot as plt

# Execution times for different problem sizes (N)
execution_times = {
    100: 460.985,
    200: 9.36751,
    300: 7.04446,
    400: 7.46323,
    500: 5.89758,
    600: 7.36825,
    700: 5.98254,
    800: 7.41803,
    900: 5.76006,
    1000: 7.53155
}

# Speedup and Parallel Efficiency for different numbers of threads
thread_data = {
    1: (3.68823, 3.68823),
    2: (9.4718, 4.7359),
    3: (7.27938, 2.42646),
    4: (9.49781, 2.37445)
}

# Graph 1: Speedup vs. Problem Size (N)
problem_sizes = list(execution_times.keys())
speedups = [execution_times[100] / execution_times[N] for N in problem_sizes]

plt.plot(problem_sizes, speedups, marker='o')
plt.xlabel('Problem Size (N)')
plt.ylabel('Speedup')
plt.title('Speedup vs. Problem Size')
plt.grid(True)
plt.show()

# Graph 2: Parallel Efficiency vs. Number of Threads
threads = list(thread_data.keys())
parallel_efficiencies = [speedups[0] / thread_data[thread][0] for thread in threads]

plt.plot(threads, parallel_efficiencies, marker='o')
plt.xlabel('Number of Threads')
plt.ylabel('Parallel Efficiency')
plt.title('Parallel Efficiency vs. Number of Threads')
plt.grid(True)
plt.show()
```
